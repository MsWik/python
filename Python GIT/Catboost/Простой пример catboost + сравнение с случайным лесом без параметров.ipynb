```python
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
from catboost import CatBoostClassifier
```

```python
ctb = CatBoostClassifier(random_seed=17)
df = pd.read_csv('bank.csv')
```

```python
train_df, y = df.drop('y', axis=1), df['y']
train_df_part, valid_df, y_train_part, y_valid = train_test_split(train_df, y, 
                                                                  test_size=.3, stratify=y, 
                                                                  random_state=17)
#разделяем выборку
```

```python
df.head()
valid_df.head()
```

```python

df.info()#смотрим содержимое
```

```python
cat_features_idx = np.where(train_df_part.dtypes == 'object')[0].tolist()#выделяем признаки типа объект
```

```python
%%time
ctb.fit(train_df_part, y_train_part,
        cat_features=cat_features_idx);
#обучаемся без параметров
```

```python
roc_auc_score(y_valid, ctb.predict_proba(valid_df)[:, 1]);#смотрим точность (0.92)
```

```python

df_no_cat, y = df.loc[:, df.dtypes != 'object'].drop('y', axis=1), df['y']
```

```python
df_no_cat_part, df_no_cat_valid, y_train_part, y_valid = train_test_split(df_no_cat, y,
                                                                            test_size=.3, 
                                                                            stratify=y,
                                                                            random_state=17)
```

```python

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=17)#подбор гиперпараметров для случайного леса
```

```python
forest = RandomForestClassifier(random_state=17,n_estimators=1000, max_depth=7)
```

```python
np.mean(cross_val_score(forest, df_no_cat_part, y_train_part, cv=skf, scoring='roc_auc'))
```

```python
forest.fit(df_no_cat_part, y_train_part)
```

```python
roc_auc_score(y_valid, forest.predict_proba(df_no_cat_valid)[:, 1])
```
